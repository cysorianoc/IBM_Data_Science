#### Creating train and test dataset
Train/Test Split involves splitting the dataset into training and testing sets respectively, which are mutually exclusive. After which, you train with the training set and test with the testing set.
This will provide a more accurate evaluation on out-of-sample accuracy because the testing dataset is not part of the dataset that have been used to train the model. Therefore, it gives us a better understanding of how well our model generalizes on new data.

We know the outcome of each data point in the testing dataset, making it great to test with! Since this data has not been used to train the model, the model has no knowledge of the outcome of these data points. So, in essence, it is truly an out-of-sample testing.

Let's split our dataset into train and test sets. Around 80% of the entire dataset will be used for training and 20% for testing. We create a mask to select random rows using the  __np.random.rand()__ function:


````python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
````

#### Train data distribution


````python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS,  color='blue')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
````

<h2 id="multiple_regression_model">Multiple Regression Model</h2>


In reality, there are multiple variables that impact the co2emission. When more than one independent variable is present, the process is called multiple linear regression. An example of multiple linear regression is predicting co2emission using the features FUELCONSUMPTION_COMB, EngineSize and Cylinders of cars. The good thing here is that multiple linear regression model is the extension of the simple linear regression model.


````python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(train[['CO2EMISSIONS']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
````

As mentioned before, __Coefficient__ and __Intercept__  are the parameters of the fitted line.
Given that it is a multiple linear regression model with 3 parameters and that the parameters are the intercept and coefficients of the hyperplane, sklearn can estimate them from our data. Scikit-learn uses plain Ordinary Least Squares method to solve this problem.

#### Ordinary Least Squares (OLS)
OLS is a method for estimating the unknown parameters in a linear regression model. OLS chooses the parameters of a linear function of a set of explanatory variables by minimizing the sum of the squares of the differences between the target dependent variable and those predicted by the linear function. In other words, it tries to minimizes the sum of squared errors (SSE) or mean squared error (MSE) between the target variable (y) and our predicted output ($\hat{y}$) over all samples in the dataset.

OLS can find the best parameters using of the following methods:
* Solving the model parameters analytically using closed-form equations
* Using an optimization algorithm (Gradient Descent, Stochastic Gradient Descent, Newton’s Method, etc.)


<h2 id="prediction">Prediction</h2>


````python
y_hat= regr.predict(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
x = np.asanyarray(test[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB']])
y = np.asanyarray(test[['CO2EMISSIONS']])
print("Mean Squared Error (MSE) : %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
````

__Explained variance regression score:__  
Let $\hat{y}$ be the estimated target output, y the corresponding (correct) target output, and Var be the Variance (the square of the standard deviation). Then the explained variance is estimated as follows:

$\texttt{explainedVariance}(y, \hat{y}) = 1 - \frac{Var\{ y - \hat{y}\}}{Var\{y\}}$  
The best possible score is 1.0, the lower values are worse.

# Example for of a prediction using two variables and plotting a prediction plane

````python
# Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Assuming the dataframe `df` is already loaded with the required data
# Independent variables are `ENGINESIZE` and `CYLINDERS`
# Dependent variable is `CO2EMISSIONS`

# Extract independent and dependent variables
x_data = df[['ENGINESIZE', 'CYLINDERS']]
y_data = df['CO2EMISSIONS']

# Split the dataset into training and testing subsets
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=1)

# Train the Multiple Linear Regression model
regr = LinearRegression()
regr.fit(x_train, y_train)

# Print model coefficients and intercept
print("\nModel Coefficients:", regr.coef_)  # Coefficients
print("Model Intercept:", regr.intercept_)  # Intercept

# Evaluate the model's performance
y_pred = regr.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print evaluation metrics
print("\nModel Evaluation Metrics:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Variance Score (R²): {r2:.2f}")

# New vehicle details: Engine size = 3.5, Cylinders = 6
new_vehicle = np.array([[3.5, 6]])  # Input data for the new vehicle
predicted_emissions = regr.predict(new_vehicle)  # Predict CO2 emissions

# Display the prediction result
print("\nPrediction Example:")
print(f"For a vehicle with Engine Size = {new_vehicle[0, 0]}, Cylinders = {new_vehicle[0, 1]}")
print(f"the predicted CO2 Emissions are {predicted_emissions[0]:.2f} g/km.")

# Visualization: 3D Scatter Plot with Prediction Plane
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of actual data points
ax.scatter(x_data['ENGINESIZE'], x_data['CYLINDERS'], y_data, c='blue', label='Actual Data', alpha=0.6)

# Create a grid for the prediction plane
engine_size_range = np.linspace(x_data['ENGINESIZE'].min(), x_data['ENGINESIZE'].max(), 10)
cylinders_range = np.linspace(x_data['CYLINDERS'].min(), x_data['CYLINDERS'].max(), 10)
engine_size_grid, cylinders_grid = np.meshgrid(engine_size_range, cylinders_range)
plane_points = np.c_[engine_size_grid.ravel(), cylinders_grid.ravel()]

# Predict CO2 emissions for the grid points
plane_predictions = regr.predict(plane_points).reshape(engine_size_grid.shape)

# Plot the prediction plane
ax.plot_surface(engine_size_grid, cylinders_grid, plane_predictions, color='red', alpha=0.5, label='Prediction Plane')

# Add labels and legend
ax.set_xlabel('Engine Size')
ax.set_ylabel('Cylinders')
ax.set_zlabel('CO2 Emissions')
ax.set_title('3D Scatter Plot with Prediction Plane')
plt.legend()
plt.show()
````

````python
# Plot alternative based on https://saturncloud.io/blog/how-to-plot-for-multiple-linear-regression-model-using-matplotlib/

# Create a 3D scatter plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the actual data points
ax.scatter(x_data['ENGINESIZE'], x_data['CYLINDERS'], y_data, c='blue', label='Actual Data', alpha=0.6)

# Fit a plane using np.linalg.lstsq
A = np.vstack([
    x_data['ENGINESIZE'],
    x_data['CYLINDERS'],
    np.ones_like(x_data['ENGINESIZE'])
]).T
plane_coef, _, _, _ = np.linalg.lstsq(A, y_data, rcond=None)

# Create a meshgrid for the plane
x_plane, y_plane = np.meshgrid(
    np.linspace(x_data['ENGINESIZE'].min(), x_data['ENGINESIZE'].max(), 10),
    np.linspace(x_data['CYLINDERS'].min(), x_data['CYLINDERS'].max(), 10)
)
z_plane = plane_coef[0] * x_plane + plane_coef[1] * y_plane + plane_coef[2]

# Add the regression plane
ax.plot_surface(x_plane, y_plane, z_plane, alpha=0.5, color='red', label='Regression Plane')

# Add labels and title
ax.set_xlabel('Engine Size')
ax.set_ylabel('Cylinders')
ax.set_zlabel('CO2 Emissions')
plt.title('Multiple Linear Regression Plane')

# Add legend
plt.legend()

# Show the plot
plt.show()
````

# Example of a prediction using three variables and metrics

````python
#Step 0: get the data
!wget -O FuelConsumption.csv https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%202/data/FuelConsumptionCo2.csv
````

````python
# Multiple Linear Regression with 3D Plane Visualization and Prediction Example

# Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Assuming the dataframe `df` is already loaded with the required data
# Independent variables are `ENGINESIZE`, `CYLINDERS`, and `FUELCONSUMPTION_COMB`
# Dependent variable is `CO2EMISSIONS`

df = pd.read_csv("FuelConsumption.csv")


# Extract independent and dependent variables
x_data = df[['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB']]
y_data = df['CO2EMISSIONS']

# Split the dataset into training and testing subsets
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=20)

# Train the Multiple Linear Regression model
regr = LinearRegression()
regr.fit(x_train, y_train)

# Print model coefficients and intercept
print("\nModel Coefficients:", regr.coef_)  # Coefficients
print("Model Intercept:", regr.intercept_)  # Intercept

# Evaluate the model's performance
y_pred = regr.predict(x_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print evaluation metrics
print("\nModel Evaluation Metrics:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Variance Score (R²): {r2:.2f}")




# Step 2: Example prediction for a new vehicle
# New vehicle details: Engine size = 3.5, Cylinders = 6, Fuel consumption combined = 10.5
new_vehicle = np.array([[3.5, 6, 10.5]])  # Input data for the new vehicle
predicted_emissions = regr.predict(new_vehicle)  # Predict CO2 emissions

# Display the prediction result
print("\nPrediction Example:")
print(f"For a vehicle with Engine Size = {new_vehicle[0, 0]}, Cylinders = {new_vehicle[0, 1]}, and Combined Fuel Consumption = {new_vehicle[0, 2]:.1f},")
print(f"the predicted CO2 Emissions are {predicted_emissions[0]:.2f} g/km.")
````
