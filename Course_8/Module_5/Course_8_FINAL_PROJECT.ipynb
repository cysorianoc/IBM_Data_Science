<br>2. Recession: A binary variable indicating recession perion; 1 means it was recession, 0 means it was normal.
<br>3. Automobile_Sales: The number of vehicles sold during the period.
<br>4. GDP: The per capita GDP value in USD.
<br>5. Unemployment_Rate: The monthly unemployment rate.
<br>6. Consumer_Confidence: A synthetic index representing consumer confidence, which can impact consumer spending and automobile purchases.

<br>10.Vehicle_Type: The type of vehicles sold; Supperminicar, Smallfamiliycar,                 Mediumfamilycar, Executivecar, Sports.
<br>11.Competition: The measure of competition in the market, such as the number of competitors or market share of major manufacturers.
<br>12.Month: Month of the observation extracted from Date..
<br>13.Year: Year of the observation extracted from Date.

### Now you want to compare the sales of different vehicle types during a recession and a non-recession period
<br>We recommend that you use the functionality of **Seaborn Library** to create this visualization


<details><summary>Click here for a hint</summary>

<p>
     To visualize sales of different vehicles during recession and non-recession periods, you can use a bar chart
        <br> You will need to group Recession, Vehicle_Type for average Automobile_Sales and then plot it<br>
    Make use of sns.barplot(x=x,y=y, data = df)</p>

````python
# Step 1: Summarizing the data by calculating the mean automobile sales for each combination of recession status and vehicle type
compare_df = df.groupby(['Recession', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].mean()
# Explanation:
# - The data is grouped by 'Recession' (indicating recession or non-recession periods) and 'Vehicle_Type' (e.g., medium family car, supermini car).
# - The 'as_index=False' ensures that the grouped columns ('Recession' and 'Vehicle_Type') remain as regular columns in the resulting DataFrame.
# - The mean of 'Automobile_Sales' is calculated for each group and stored in a new DataFrame called 'compare_df'.
# - The resulting DataFrame contains the average sales for each vehicle type, split into recession and non-recession periods.

# Step 2: Creating a bar plot to compare automobile sales during recession and non-recession periods for different vehicle types
sns.barplot(x='Recession', y='Automobile_Sales', hue='Vehicle_Type', data=compare_df)
# Explanation:
# - A bar plot is created using Seaborn's `barplot` function.
# - The x-axis ('Recession') represents whether the period is a recession (1) or non-recession (0).
# - The y-axis ('Automobile_Sales') shows the average sales for each vehicle type.
# - The 'hue' parameter groups the bars by 'Vehicle_Type', differentiating them by color.
# - The input data for the plot is the summarized DataFrame, 'compare_df'.

# Step 3: Customizing the x-axis labels to make them more descriptive
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])
# Explanation:
# - The x-axis ticks are customized to replace the numeric recession indicators (0 and 1) with more descriptive labels: 'Non-Recession' and 'Recession'.
# - The `ticks` argument specifies the tick positions (aligned with the values of 'Recession').
# - The `labels` argument provides the corresponding text for these tick positions.

# Step 4: Adding a label for the x-axis
plt.xlabel('Recession')
# Explanation:
# - The x-axis label is set to "Recession" to clarify that the x-axis represents recession status.

# Step 5: Adding a label for the y-axis
plt.ylabel('Automobile sales (mean)')
# Explanation:
# - The y-axis label is set to "Automobile sales (mean)" to indicate that the y-axis represents the average automobile sales.

# Step 6: Adding a title to the plot
plt.title('Vehicle-Wise Sales during Recession and Non-Recession Period')
# Explanation:
# - A title is added to summarize the purpose of the plot.
# - The title highlights that the visualization compares vehicle-wise sales during recession and non-recession periods.
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


````python
df.head()
````

<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>

````python
# Filter the DataFrame to include only data from non-recession periods (Recession = 0)
df_nonrecession = df[df['Recession'] == 0]
# Explanation: The above line filters the original DataFrame 'df' to create a new DataFrame 'df_nonrecession' that
# only includes rows where the 'Recession' column equals 0, representing non-recession periods.

# Create a scatter plot using Seaborn
sns.scatterplot(data=df_nonrecession, x='Month', y='Automobile_Sales', size='Seasonality_Weight', hue='Seasonality_Weight')
# Explanation: This line creates a scatter plot using Seaborn:
# - 'data=df_nonrecession' specifies the DataFrame to be used for the plot (only non-recession data).
# - 'x="Month"' represents the month data on the x-axis.
# - 'y="Automobile_Sales"' represents the automobile sales data on the y-axis.
# - 'size="Seasonality_Weight"' adjusts the size of each point based on the 'Seasonality_Weight' column.
# - 'hue="Seasonality_Weight"' assigns colors to the points based on the 'Seasonality_Weight' column,
#   allowing the viewer to visually assess how seasonality impacts sales.

# Label the x-axis as 'Month'
plt.xlabel('Month')
# Explanation: The x-axis is labeled as 'Month' to indicate that the x-axis represents the months of the year.

# Label the y-axis as 'Automobile_Sales'
plt.ylabel('Automobile_Sales')
# Explanation: The y-axis is labeled as 'Automobile_Sales' to indicate that the sales data is plotted along this axis.

# Add a title to the plot
plt.title('Seasonality impact on Automobile Sales')
# Explanation: The title of the plot is set to 'Seasonality impact on Automobile Sales', providing context for the viewer.
# It tells the viewer that the plot demonstrates the relationship between automobile sales and seasonality during non-recession periods.

# Display the plot
plt.show()
# Explanation: This line displays the plot on the screen, rendering the visual representation of the data.
````

### TASK 1.8: Develop a pie chart to display the total Advertisement expenditure for each vehicle type during recession period.<br>
Can we observe the share of each vehicle type in total expenditure during recessions?


<details><summary>Click here for a hint</summary>

<p>
     You will be required to group vehicle type for sum of advertisement expenditure.
    <br> the plot a pie with the data, May include relevant labels
    <br>title this plot as  - Share of Each Vehicle Type in Total Expenditure during Recessions</p>
</details>


````python
# Filter the dataset for the recession period
Rdata = df[df['Recession'] == 1]  # Extract data where 'Recession' column is 1 (indicating recession period)

# Group the data by 'Vehicle_Type' and calculate the sum of 'Advertising_Expenditure' for each vehicle type during the recession
VTexpenditure = Rdata.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()
# Explanation: The `groupby('Vehicle_Type')` groups the data by different vehicle types, and `sum()` calculates the total advertising expenditure for each vehicle type during the recession period.

# Create a figure for the pie chart with specified size
plt.figure(figsize=(8, 6))  # Define the figure size to ensure it's appropriately scaled for viewing

# Set the labels and sizes for the pie chart
labels = VTexpenditure.index  # The vehicle types are used as labels for the pie chart slices (these are the index values from the grouped data)
sizes = VTexpenditure.values  # The corresponding total advertising expenditure values for each vehicle type

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
# Explanation:
# 'sizes' are the total advertising expenditure values for each vehicle type.
# 'labels' are the vehicle types which will appear as labels on the pie chart slices.
# 'autopct='%1.1f%%'' shows the percentage of the total expenditure for each vehicle type in the pie chart.
# 'startangle=90' rotates the chart so the first slice starts from the top (90 degrees).

# Add a title to the pie chart
plt.title('Share of Each Vehicle Type in Total Expenditure during Recessions')
# Explanation: This title clarifies that the pie chart is showing the proportion of total advertising expenditure attributed to each vehicle type during recession periods.

# Display the pie chart
plt.show()  # This renders and displays the pie chart

````
