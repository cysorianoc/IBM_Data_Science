The data set to be used is available on the link below.


The functions below will download the dataset into your browser:


````python
# From the original code, those lines did not work
#because pyodide is not a module that's natively available in a Colab notebook environment.
#Pyodide is typically used in environments that run Python in the browser, such as web-based environments like JupyterLite,
#and is not intended for use in Google Colab or other standard Python environments.
from pyodide.http import pyfetch

async def download(url, filename):
    response = await pyfetch(url)
    if response.status == 200:
        with open(filename, "wb") as f:
            f.write(await response.bytes())
````

````python
# Alternativaly we can request an HTTP to download the file from the project URL
import requests

def download(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, 'wb') as f:
            f.write(response.content)




url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-Coursera/laptop_pricing_dataset_base.csv"
file_name = 'laptops.csv'
download(url, file_name)
````

Now, we can convert the .csv file into a dataframe:


````python
df = pd.read_csv(file_name)
````
